<h1><b>For e Listas</h1></b>

O uso do for como contador pode ser muito útil para sincronizar várias listas. 
Supondo que temos listas para armazenar vários dados diferentes de várias pessoas. Se as listas tiverem ordenada, onde o índice 0 de cada lista é o dado de uma pessoa, podemos recuperar os dados dos cadastros facilmente.

Para isso usamos o for como um contador e usamos a variável i para sincronizar todas as listas.

```
# Faça um programa de cadastro de pessoas que receba 5 nomes, idades e e-mails e
# salve cada um em uma lista. Depois mostre os dados dos clientes.

################################################################################
#                          Captação de dados                                   #
################################################################################

# iniciando listas vazias para capturar dados dos clientes
lista_nomes = [ ]
lista_idades = [ ]
lista_email = [ ]

# Usando o for a nosso favor, para cadastrar novos clientes
for i in range(3):
    # Entrada dos dados
    nome = input(f'Digite o nome {1+i}: ' )
    idade = input(f'Digite o idade {1+i}: ' )
    email = input(f'Digite o email {1+i}: ' )
    # Adicionando os dados nas listas.
    lista_nomes.append( nome )
    lista_idades.append( idade )
    lista_email.append( email )
    print('\n')

# Até aqui tudo bem, foi o mais do mesmo!

################################################################################
#                      Apresentação dos resultados                             #
################################################################################

# Para sincronizarmos a lista, devemos pegar uma e medir o tamanho dela.
# Como eu sei que todas tem o mesmo tamanho, qualquer uma das listas de cadastro
# já é o suficiente.

tamanho = len( lista_nomes )

# Vamos usar o for como contador e colocar a variável tamanho dentro do range()
# Assim ela irá fazer o loop na medida certa para pegar-mos todos os dados 
# cadastrados.
for i in range( tamanho ): 
    # Printamos os resultados.
    print( f'''
nome: { lista_nomes[ i ] }
Idade: { lista_idades[ i ] }
E-mail: { lista_email[ i ] } 
''' )
```

<b>Exercício 01</b>

Crie um programa que cadastre 5 clientes. 

Cada cliente possui: Nome, sexo, Telefone<br>
(Guarde estes dados em listas separadas).

Depois mostre os dados cadastrados no seguinte formato:



\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>
Relatório de clientes cadastrados <br>
\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>
Nome: <br>
Sexo:<br>
Telefone:<br>
\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>

```
from abc import ABC


class Pessoa(ABC):

    def __init__(self, nome, sexo):
        self.__nome = nome
        self.__sexo = sexo

    @property
    def nome(self):
        return self.__nome

    @property
    def sexo(self):
        return self.__sexo


class Cliente(Pessoa):

    def __init__(self, nome, sexo, telefone):
        super().__init__(nome, sexo)
        self.__telefone = telefone

    @property
    def telefone(self):
        return self.__telefone


if __name__ == "__main__":

    lista_clientes = []
    for i in range(5):
        nome = input(f"Digite o nome do {i + 1} cliente: ")
        sexo = input(f"Digite o sexo do {i + 1} cliente: ")
        telefone = input(f"Digite o telefone do {i + 1} cliente: ")
        lista_clientes.append(Cliente(nome, sexo, telefone))

    for cliente in lista_clientes:
        print()
        print("*********************************\nRelatório de clientes cadastrados\n*********************************")
        print(f"Nome: {cliente.nome}")
        print(f"Sexo: {cliente.sexo}")
        print(f"Telefone: {cliente.telefone}")
        print("*********************************")
```

<b>ATENÇÃO!!!!<br>
O id é um número atribuido automáticamente! O cliente não escolhe o número. O seu programa deve fazer o cadastro deste id automaticamente.</b><br><br><br>


Com isso, crie um cadastro de clientes que receba o id, nome e idade. Depois mostre os dados dos clientes individualmente.
(cadastre no minimo 4 clientes)


```
from abc import ABC


class Pessoa(ABC):

    def __init__(self, nome, sexo):
        self.__nome = nome
        self.__sexo = sexo

    @property
    def nome(self):
        return self.__nome

    @property
    def sexo(self):
        return self.__sexo


class Cliente(Pessoa):

    cliente_id = 0

    def __init__(self, nome, sexo, telefone):
        super().__init__(nome, sexo)
        self.__id = Cliente.cliente_id + 1
        self.__telefone = telefone
        Cliente.cliente_id = self.__id

    @property
    def telefone(self):
        return self.__telefone

    @property
    def id(self):
        return self.__id


if __name__ == "__main__":

    lista_clientes = []
    for i in range(5):
        nome = input(f"Digite o nome do {i + 1} cliente: ")
        sexo = input(f"Digite o sexo do {i + 1} cliente: ")
        telefone = input(f"Digite o telefone do {i + 1} cliente: ")
        lista_clientes.append(Cliente(nome, sexo, telefone))

    for cliente in lista_clientes:
        print()
        print("*********************************\nRelatório de clientes cadastrados\n*********************************")
        print(f"Id: {cliente.id}")
        print(f"Nome: {cliente.nome}")
        print(f"Sexo: {cliente.sexo}")
        print(f"Telefone: {cliente.telefone}")
        print("*********************************")
```

<b>Exercício 03</b>

3.1) Crie um programa que cadastre o id, nome, sexo e a idade de 5 clientes.

3.2) Depois mostre os dados dos 5 clientes.

3.3) Peça para que o usuário escolha um id de um cliente e mostre as informações deste cliente.

3.4) Crie um filtro que ao digitar um id de um cliente mostre as seguintes informações:<br>
- Para menores de 17 anos: Entrada Proibida
- Para maiores de 17 anos: Entrada Liberada
- Para o sexo Feminino: 50% de desconto
- Para o sexo Masculino: 5% de desconto



```
from abc import ABC
from typing import Union

class Pessoa(ABC):

    def __init__(self, nome, sexo, idade):
        self.__nome = nome
        self.__sexo = sexo
        self.__idade = idade

    @property
    def nome(self):
        return self.__nome

    @property
    def sexo(self):
        return self.__sexo

    @property
    def idade(self):
        return self.__idade


class Cliente(Pessoa):

    cliente_id = 0

    def __init__(self, nome, sexo, idade, telefone):
        super().__init__(nome, sexo, idade)
        self.__id = Cliente.cliente_id + 1
        self.__telefone = telefone
        Cliente.cliente_id = self.__id

    @property
    def telefone(self):
        return self.__telefone

    @property
    def id(self):
        return self.__id

    def filtro_idade(self):
        if self.idade < 17:
            return "Entrada Proibida"
        else:
            return "Entrada Liberada"

    def filtro_sexo(self):
        if self.sexo == "f" or self.sexo == "feminino":
            return "50% de desconto"
        elif self.sexo == "m" or self.sexo == "masculino":
            return "5% de desconto"
        else:
            return "Sexo Inválido"


def mostra_dados(obj: object, msg: str, **kwargs) -> None:
    """Saída de dados"""
    print()
    print(f"*********************************\n{msg}\n*********************************")
    print(f"Id: {obj.id}")
    print(f"Nome: {obj.nome}")
    print(f"Sexo: {obj.sexo}")
    print(f"Idade: {obj.idade}")
    print(f"Telefone: {obj.telefone}")
    if "filtros" in kwargs and kwargs["filtros"]:
        print(f"Status: {obj.filtro_idade()}")
        print(f"Desconto: {obj.filtro_sexo()}")
    print("*********************************")


def inpput(mensagem: str, tipo: (int, float) = float) -> Union[int, float, None]:
    """Input com validação de dado (Int ou Float)"""
    if tipo is float or tipo is int:
        while True:
            try:
                valor = tipo(input(mensagem))
                return valor
            except ValueError:
                print(f"O valor precisa ser do tipo {tipo}")
    else:
        raise TypeError(f"O tipo {tipo} é inválido")


if __name__ == "__main__":

    # Cadastrando 5 clientes
    lista_clientes = []
    for i in range(5):
        nome = input(f"Digite o nome do {i + 1} cliente: ")
        sexo = input(f"Digite o sexo do {i + 1} cliente (m | f): ").lower()
        idade = inpput(f"Digite a idade do {i + 1} cliente: ", int)
        telefone = input(f"Digite o telefone do {i + 1} cliente: ")
        lista_clientes.append(Cliente(nome, sexo, idade, telefone))

    # Exibindo os 5 clientes
    for cliente in lista_clientes:
        mostra_dados(cliente, msg="Relatório de clientes cadastrados")

    # Escolhendo cliente pelo ID
    id_escolhido = inpput("Escolha um ID: ", int)

    # Utilizando o filter para obter um Iterator com o objeto que possui o id escolhido
    filtro = filter(lambda x: x.id == id_escolhido, lista_clientes)
    # Como apenas um objeto é retornado basta utilizar a função next() para obter este objeto
    try:
        obj_filter = next(filtro)
        mostra_dados(obj_filter, "Relatório do cliente (filter)", filtros=True)
    except StopIteration:
        print("ID inválido")

    # Fazendo a mesma coisa utilizando Generator
    generator = (x for x in lista_clientes if x.id == id_escolhido)
    try:
        obj_generator = next(generator)
        mostra_dados(obj_filter, "Relatório do cliente (generator)", filtros=True)
    except StopIteration:
        print("ID Inválido")

    # Fazendo a mesma coisa utilizando o Loop For
    retorno = False
    for cliente in lista_clientes:
        if cliente.id == id_escolhido:
            mostra_dados(cliente, "Relatório do cliente (loop for)", filtros=True)
            retorno = True
            break

    if not retorno:
        print("ID Inválido")
```
